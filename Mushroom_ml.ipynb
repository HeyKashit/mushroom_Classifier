```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
Data= pd.read_csv('https://raw.githubusercontent.com/HeyKashit/mushroom_ml_eda/main/new_data.csv')
```

```python
Data.head()
```

```python
X=Data.drop(columns='class')
Y=Data['class']
```

```python
train,train_val,test,test_val=train_test_split(X,Y,test_size=0.4,random_state=477)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(train,test,test_size=0.2,random_state=577)
```

 Using knn as 1 base model

```python
knn=KNeighborsClassifier ()
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

Using svm as 2 base model

```python
svm=SVC()
svm.fit(X_train,Y_train)
```

```python
svm.score(X_test,Y_test)
```

Now prdiciting the values using Svm and knn model

```python
Pre1=knn.predict(train_val)
Pre2=svm.predict(train_val)
```

 Now making a stack of both base model predictions using numpy

```python
Pre=np.column_stack((Pre1,Pre2))
```

```python
 
predict_test1 = knn.predict(X_test)
predict_test2 = svm.predict(X_test)
```

```python
 predict_test = np.column_stack((predict_test1,predict_test2))
predict_test
```

```python
random= RandomForestClassifier ()
random.fit(Pre,test_val)
```

```python
random.score(predict_test,Y_test)
```

Now the final accuracy of our model is 81.19%
