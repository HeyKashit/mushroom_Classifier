```python
import pandas as pd
import numpy as np
```

```python
Data=pd.read_csv('new_data.csv')
```

```python
Data.head()
```

```python
Data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Data.head()
```

```python
x=Data.drop(['class'],axis=1)
y=Data['class']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.1,random_state=117)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,ELU
```

```python
classifier = Sequential()

classifier.add(Dense(units = 24, kernel_initializer= 'he_uniform',activation='ELU',input_dim = 94))
classifier.add(Dense(units = 12, kernel_initializer = 'he_uniform',activation='ELU'))
classifier.add(Dense(units = 30, kernel_initializer = 'he_uniform',activation='ELU'))
classifier.add(Dense(units =5, kernel_initializer = 'he_uniform',activation='ELU'))
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))

classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])
model_history=classifier.fit(x_train, y_train,validation_split=0.23, batch_size = 10)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```
